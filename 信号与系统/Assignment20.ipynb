## (a)

当初始条件均为 0 且输入  
$$
x(t)=e^{-4t}u(t)\quad \Longrightarrow\quad X(s)=\frac{1}{s+4},
$$
对方程  
$$
y'''(t)+6y''(t)+11y'(t)+6y(t)=x(t)
$$
作单边拉普拉斯变换，由于零初始条件，变换后为  
$$
\left[s^3+6s^2+11s+6\right]Y(s)=\frac{1}{s+4}.
$$
注意因式分解  
$$
s^3+6s^2+11s+6=(s+1)(s+2)(s+3),
$$
因此  
$$
Y(s)=\frac{1}{(s+4)(s+1)(s+2)(s+3)}.
$$
接下来对 $Y(s)$ 进行部分分式展开，设
$$
Y(s)=\frac{A}{s+4}+\frac{B}{s+1}+\frac{C}{s+2}+\frac{D}{s+3},
$$
利用留数法可求得：
  
- $s=-4$：  
  $$
  A=\lim_{s\to-4}(s+4)Y(s)=\frac{1}{(-3)(-2)(-1)}=-\frac{1}{6}.
  $$
- $s=-1$：  
  $$
  B=\lim_{s\to-1}(s+1)Y(s)=\frac{1}{(3)(1)(2)}=\frac{1}{6}.
  $$
- $s=-2$：  
  $$
  C=\lim_{s\to-2}(s+2)Y(s)=\frac{1}{(2)(-1)(1)}=-\frac{1}{2}.
  $$
- $s=-3$：  
  $$
  D=\lim_{s\to-3}(s+3)Y(s)=\frac{1}{(1)(-2)(-1)}=\frac{1}{2}.
  $$
  
因此  
$$
y_{\text{zs}}(t)=\mathcal{L}^{-1}\{Y(s)\}=-\frac{1}{6}e^{-4t}u(t)+\frac{1}{6}e^{-t}u(t)-\frac{1}{2}e^{-2t}u(t)+\frac{1}{2}e^{-3t}u(t).
$$

## (b) 零输入响应

设输入 $x(t)=0$，初始条件为  
$$
y(0^-)=1,\quad y'(0^-)=-1,\quad y''(0^-)=1.
$$
对方程作单边拉普拉斯变换时，需引入初始条件项：
$$
\begin{aligned}
\mathcal{L}\{y'''(t)\} &= s^3Y(s)-s^2y(0^-)-sy'(0^-)-y''(0^-),\\[1mm]
\mathcal{L}\{y''(t)\} &= s^2Y(s)-sy(0^-)-y'(0^-),\\[1mm]
\mathcal{L}\{y'(t)\} &= sY(s)-y(0^-).
\end{aligned}
$$
因此变换后方程为  
$$
\begin{aligned}
&s^3Y(s)-s^2(1)-s(-1)-1 \\
&\quad +6\Bigl[s^2Y(s)-s(1)-(-1)\Bigr] \\
&\quad +11\Bigl[sY(s)-1\Bigr] + 6Y(s)=0.
\end{aligned}
$$
整理 $Y(s)$ 项和初始条件项，注意 $s^3+6s^2+11s+6=(s+1)(s+2)(s+3)$ 
最终可化简得到  
$$
Y(s)=\frac{1}{s+1}.
$$
因此  
$$
y_{\text{zi}}(t)=\mathcal{L}^{-1}\{Y(s)\}=e^{-t}u(t).
$$

X(z)=\sum_{n=-\infty}^{+\infty}\delta[n+5]\,z^{-n}
=\;z^{-(-5)}=z^{5}.

**收敛域（ROC）**

- 对于有限项序列（有限长度的 z 变换），级数在全平面收敛。  
- $\textbf{ROC:}$ 全部复平面。

**傅立叶变换**

- 当取 $z=e^{j\omega}$ 时，$X(e^{j\omega})=e^{j5\omega}$ 明显存在且连续。  
- 因此该序列的离散傅立叶变换存在。

### (g) 
$$
x[n]=2^n u[-n] + \left(\tfrac{1}{4}\right)^n u[n-1]
$$

分成两部分求 z 变换。

#### 第一部分： $x_1[n]=2^n\, u[-n]$

$$
X_1(z)=\sum_{n=-\infty}^{0}2^n\,z^{-n}.
$$
令 $m=-n$（当 $n\le0$ 时 $m\ge0$ ），得
$$
X_1(z)=\sum_{m=0}^{\infty}2^{-m}\,z^{m}
=\sum_{m=0}^{\infty}\left(\frac{z}{2}\right)^{m}
=\frac{1}{1-\frac{z}{2}},
$$
$$
\left|\frac{z}{2}\right|<1\quad\Longrightarrow\quad |z|<2.
$$
因此，$X_1(z)=\dfrac{1}{1-z/2}$，其唯一极点在 $z=2$。

#### 第二部分： $x_2[n]=\left(\tfrac{1}{4}\right)^n\, u[n-1]$

$$
X_2(z)=\sum_{n=1}^{\infty}\left(\frac{1}{4}\right)^n\,z^{-n}
=\sum_{n=1}^{\infty}\left(\frac{1/4}{z}\right)^n.
$$
该级数为几何级数，其和为
$$
X_2(z)=\frac{\frac{1}{4z^{-1}}}{1-\frac{1}{4z^{-1}}}
=\frac{(1/4)z^{-1}}{1-\frac{1}{4}z^{-1}}.
$$

$$
X_2(z)=\frac{1/4}{z-\frac{1}{4}},
$$

$$
\left|\frac{1}{4z}\right| < 1 \quad\Longrightarrow\quad |z|>\frac{1}{4}.
$$
因此，$X_2(z)$ 的唯一极点在 $z=\tfrac{1}{4}$。

总 z 变换为
$$
X(z)=X_1(z)+X_2(z)
=\frac{1}{1-\frac{z}{2}}+\frac{1/4}{z-\frac{1}{4}}.
$$

**收敛域（ROC）**

因此，两部分共同 ROC 为
$$
\frac{1}{4} < |z| < 2.
$$

**傅立叶变换**

- 离散傅立叶变换是取 $z=e^{j\omega}$ 的值。  
- 由 ROC $\frac{1}{4}<|z|<2$ 可知单位圆 $|z|=1$ 属于 ROC。  
- 因此序列的傅立叶变换存在。

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建包含两个子图的图形
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# -----------------------------
# (a) 序列 X(z)=z^5 的零－极点图
zeros_a = np.zeros(5)       # 5个零在 z=0
poles_a = np.array([])      # 没有有限极点

axs[0].plot(np.real(zeros_a), np.imag(zeros_a), 'bo', markersize=10, label='Zeros')
if poles_a.size > 0:
    axs[0].plot(np.real(poles_a), np.imag(poles_a), 'rx', markersize=10, label='Poles')
axs[0].axhline(0, color='black', lw=0.5)
axs[0].axvline(0, color='black', lw=0.5)
axs[0].set_title("Zero-Pole Plot for X(z)=z^5")
axs[0].set_xlabel("Real Part")
axs[0].set_ylabel("Imaginary Part")
axs[0].legend()
axs[0].grid(True)

# -----------------------------
# (g) 部分的零－极点图
# 合成表达式: X(z)= -(7/4)*z / ((z-2)(z-1/4))
# 零点：z = 0 (单零)
# 极点：z = 2 和 z = 1/4
zeros_g = np.array([0])
poles_g = np.array([2, 0.25])

axs[1].plot(np.real(zeros_g), np.imag(zeros_g), 'bo', markersize=10, label='Zeros')
axs[1].plot(np.real(poles_g), np.imag(poles_g), 'rx', markersize=10, label='Poles')
axs[1].axhline(0, color='black', lw=0.5)
axs[1].axvline(0, color='black', lw=0.5)
# 绘制单位圆 (用于验证单位圆属于 ROC: 1/4<|z|<2)
theta = np.linspace(0, 2*np.pi, 400)
axs[1].plot(np.cos(theta), np.sin(theta), 'g--', label='Unit Circle')

axs[1].set_title("Zero-Pole Plot for X(z) in part (g)")
axs[1].set_xlabel("Real Part")
axs[1].set_ylabel("Imaginary Part")
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
