# 5.21(e)

计算下列信号的离散傅立叶变换

$$
x[n] = \left(\frac{1}{2}\right)^{|n|}\cos\left(\frac{\pi}{8}(n-1) \right)
$$

## Answer

令 $x[n] = x_1[n]\cdot x_2[n - 1], \quad x_1[n] = \left(\frac{1}{2}\right)^{|n|}, \quad x_2[n - 1] = \cos\left( \frac{\pi}{8}(n-1) \right)$

则，

$$
\begin{aligned}
    X_1(e^{j\omega}) &= \sum_{n = -\infty}^\infty x[n]e^{-j\omega n} \\
    &= \sum_{n = 0}^\infty \left( \frac{1}{2}e^{-j\omega} \right)^n + \sum_{m = 1}^\infty \left( \frac{1}{2}e^{j\omega} \right)^m \\
    &= \frac{1}{1 - \frac{1}{2}e^{-j\omega}} + \frac{\frac{1}{2}e^{j\omega}}{1 - \frac{1}{2}e^{j\omega}} \\
    &= \frac{3}{5 - 4\cos \omega}
\end{aligned}
$$

我们有

$$
x_2[n] = \cos\Bigl(\frac{\pi}{8}n\Bigr)
= \frac{1}{2}\Bigl(e^{j\frac{\pi}{8}n}+e^{-j\frac{\pi}{8}n}\Bigr).
$$

因此（利用 DTFT 中冲激函数对），
$$
X_2(e^{j\omega}) = \pi\Bigl[\delta\Bigl(\omega-\frac{\pi}{8}\Bigr)
+\delta\Bigl(\omega+\frac{\pi}{8}\Bigr)\Bigr]\,.
$$

利用时移性质：
$$
\mathcal{DTFT}\{x_2[n-1]\} = e^{-j\omega\cdot1}\,X_2(e^{j\omega})
= e^{-j\omega}\,\pi\Bigl[\delta\Bigl(\omega-\frac{\pi}{8}\Bigr)
+\delta\Bigl(\omega+\frac{\pi}{8}\Bigr)\Bigr]\,.
$$

因为
$$
x[n]=x_1[n]\cdot x_2[n-1],
$$
所以它的 DTFT 为
$$
X(e^{j\omega}) = \frac{1}{2\pi}\Bigl[X_1(e^{j\omega}) * \Bigl\{e^{-j\omega}\,\pi\Bigl[\delta(\omega-\frac{\pi}{8})
+\delta(\omega+\frac{\pi}{8})\Bigr]\Bigr\}\Bigr]\,.
$$

也就是说，
$$
X(e^{j\omega}) = \frac{1}{2}\int_{-\pi}^{\pi}X_1(e^{j\theta})\, e^{-j(\omega-\theta)}\,
\Bigl[\delta((\omega-\theta)-\frac{\pi}{8})
+\delta((\omega-\theta)+\frac{\pi}{8})\Bigr]d\theta.
$$

利用冲激函数的抽样性质，积分将“采样”到两点：
- 当 $\omega-\theta = \frac{\pi}{8}$ 时，即 $\theta = \omega-\frac{\pi}{8}$；
- 当 $\omega-\theta = -\frac{\pi}{8}$ 时，即 $\theta = \omega+\frac{\pi}{8}$。

于是
$$
X(e^{j\omega}) = \frac{1}{2}\Biggl\{ e^{-j\frac{\pi}{8}}\,X_1\Bigl(e^{j(\omega-\frac{\pi}{8})}\Bigr) + e^{j\frac{\pi}{8}}\,X_1\Bigl(e^{j(\omega+\frac{\pi}{8})}\Bigr)\Biggr\}\,.
$$

将 $X_1(e^{j\theta}) = \frac{3}{5-4\cos\theta}$ 代入，得

$$
\boxed{
X(e^{j\omega}) = \frac{1}{2}\Biggl[ e^{-j\frac{\pi}{8}}\frac{3}{5-4\cos\Bigl(\omega-\frac{\pi}{8}\Bigr)} + e^{j\frac{\pi}{8}}\frac{3}{5-4\cos\Bigl(\omega+\frac{\pi}{8}\Bigr)}\Biggr]\,.
}
$$

# 5.22(a)

求相应于下列傅立叶变换的信号

$$
X(e^{j\omega}) = \begin{cases}
1, & \frac{\pi}{4} \le |\omega \le \frac{3\pi} 4 \\
0, & \frac{3\pi}{4} \le |\omega| \le \pi, 0\le |\omega| < \frac \pi 4
\end{cases}
$$

## Answer

观察给定的 $X(e^{j\omega})$ 可看作两个方波函数的差：

- 令
  $$
  X_1(e^{j\omega}) = \begin{cases}
  1, & |\omega| \le \frac{3\pi}{4},\\[1mm]
  0, & \text{otherwise},
  \end{cases}
  $$
- 令
  $$
  X_2(e^{j\omega}) = \begin{cases}
  1, & |\omega| \le \frac{\pi}{4},\\[1mm]
  0, & \text{otherwise}.
  \end{cases}
  $$
  
则显然有
$$
X(e^{j\omega}) = X_1(e^{j\omega}) - X_2(e^{j\omega}).
$$

利用标准结论：对于
$$
\tilde x[n] = \frac{1}{2\pi}\int_{-W}^{W} e^{j\omega n} d\omega,
$$
计算可得
$$
\tilde x[n] = \frac{\sin(W n)}{\pi n},\quad n\ne0,\quad \tilde x[0] = \frac{W}{\pi}.
$$

因此：

- 对于 $X_1(e^{j\omega})$（这里 $W=\frac{3\pi}{4}$）有
  $$
  x_1[n] = \frac{\sin\Bigl(\frac{3\pi n}{4}\Bigr)}{\pi n}, \quad n\ne 0,\quad x_1[0] = \frac{3\pi/4}{\pi} = \frac{3}{4}\,.
  $$
- 对于 $X_2(e^{j\omega})$（这里 $W=\frac{\pi}{4}$）有
  $$
  x_2[n] = \frac{\sin\Bigl(\frac{\pi n}{4}\Bigr)}{\pi n}, \quad n\ne 0,\quad x_2[0] = \frac{\pi/4}{\pi} = \frac{1}{4}\,.
  $$

由频域的线性叠加（差分）关系，其对应的时域信号为

$$
x[n] = x_1[n] - x_2[n].
$$

即，对于 $n\ne 0$ 有

$$
x[n] = \frac{\sin\Bigl(\frac{3\pi n}{4}\Bigr)- \sin\Bigl(\frac{\pi n}{4}\Bigr)}{\pi n},
$$

而 $n=0$ 时，

$$
x[0] = x_1[0]-x_2[0] = \frac{3}{4} - \frac{1}{4} = \frac{1}{2}\,.
$$

综上，

$$
\boxed{
x[n] = \begin{cases}
\frac{1}{2}, & n=0, \\[1mm]
\frac{1}{\pi n}\Bigl[\sin\Bigl(\frac{3\pi n}{4}\Bigr)- \sin\Bigl(\frac{\pi n}{4}\Bigr)\Bigr], & n\neq0\,.
\end{cases}
}
$$
